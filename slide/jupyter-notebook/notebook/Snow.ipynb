```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plot

csv = 'monthly_snow.csv' # 1962/01 - 2018/02 (昭和37年 - 平成30年)
df = pd.read_csv(csv,
                 skiprows=6,
                 index_col=0,
                 parse_dates=True,
                 names=['date', 'fukui'],
                 usecols=[0,1]
                )

df.plot(figsize=(16, 6), alpha=0.5, ylim=[0,220])
```

```python
csv = '1963_daily.csv'
df1963 = pd.read_csv(csv,
                 skiprows=6,
                 index_col=0,
                 parse_dates=True,
                 names=['date', 'daily_1963', 'total1_963'],# 'c', 'total', 'd', 'e', 'f'],
                 usecols=[0,1,5]
                )
df1963.plot(figsize=(16, 6), alpha=0.5, grid=True, ylim=[0,220])
```
